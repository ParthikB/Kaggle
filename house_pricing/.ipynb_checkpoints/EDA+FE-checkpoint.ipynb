```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('database/train.csv', index_col=0)
data.head()
```

# NaN Values

```python
def plot_nan_values(df):
    global nan_features
    nan_features = df.columns[df.isnull().sum() > 1]
    
    global nan_percentage_of_features
    nan_percentage_of_features = df[nan_features].isnull().sum()/len(df)*100
    nan_percentage_of_features.plot(kind='bar', figsize=(20, 5))

    plt.ylabel('NaN Percentage')
    plt.xticks(rotation=50)
    plt.grid(1)
    
    
plot_nan_values(data)
```

As we can see, there are 5 features with Nan Percentage around and above 50. 

**PROCESSING** : But a few(4) features have ridiculously high Nan Percentages, above 80. We'll be completely dropping them.

```python
THRESHOLD = 80

features_higher_than_threshold_nan = nan_percentage_of_features[nan_percentage_of_features>THRESHOLD]
data = data.drop(columns=features_higher_than_threshold_nan.index)
```

```python
plot_nan_values(data)
```

Now, we'll see the relationship between FireplaceQu/LotFrontage and SalePrice

```python
df = data.copy()
df['FireplaceQu'] = data['FireplaceQu'].fillna('missing')

plt.figure(figsize=(20, 8))
sns.scatterplot(x=data.index, y='SalePrice', data=df, hue='FireplaceQu')
```

```python
data['LotFrontage'].plot(kind='box', vert=False)
```

**PROCESSING** : As we can see..

* FireplaceQu : most of the NaN values lies within the range 100k-200k. We can use this information, thus, converting the NaN Labels into a category
* LotFrontage : there are a lot of outliers in LotFrontage, so, replacing the NaN with MEDIAN.

```python
data['FireplaceQu'].fillna('missing', inplace=True)
data['LotFrontage'].fillna(data['LotFrontage'].median(), inplace=True)

plot_nan_values(data)
```

Now, our Nan values are under 5% per category, now, we'll check for RARE categories and merge them with the NaN

```python
cat_features = [f for f in data.columns if data[f].dtype=='O']
num_features = [f for f in data.columns if data[f].dtype!='O' and f not in ['Id', 'SalePrice']]

print('Total number of Categorical Features :', len(cat_features))
print('Total number of Numerical Features   :', len(num_features))
```

```python
THRESHOLD = 5

df = data.copy()

for f in cat_features:
    df[f].fillna('other', inplace=True)
    frequency_percent = df.groupby(f)[f].count()/len(data) * 100
#     print(frequency_percent, '\n')
    for val in frequency_percent[frequency_percent<THRESHOLD].index:
        df[f] = np.where(df[f] == val, 'other', df[f])
        
data = df.copy()
```

```python
num_nan_features = [f for f in num_features if df[f].isnull().sum() >= 1]

plt.figure(figsize=(25, 25), dpi=100)
for idx, f in enumerate(num_nan_features):
    plt.subplot(10, 4, idx+1)
    df[f].plot(kind='box', vert=False)
    plt.yticks(rotation=90)
```

Clearly we can see that MasVnrArea has a lot of outliers and GarageYrBlt does not.
Thus..

```python
data['MasVnrArea'].fillna(data['MasVnrArea'].median(), inplace=True)
data['GarageYrBlt'].fillna(data['GarageYrBlt'].mean(), inplace=True)

print('Total Number of NaN Values in our dataset now :', df.isnull().sum().sum())
```

## Types of Numerical Features:
    
* Temporal
* Discrete
* Continuous

Let's Split the num_features in the mentioned categories first

#### Temporal Features

```python
# Manually Selecting the Temporal Features
temporal_num_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.grid(1)
```

Well that's odd, the price should go up due to inflation as the time passes. But that's not the pattern here.

Let's see the pattern between other temporal variables with our target feature.

```python
plt.figure(figsize=(30, 5))
for idx, f in enumerate(temporal_num_features):
    df = data.copy()
    
    df[f] = df['YrSold'] - df[f]
    
    plt.subplot(1, 4, idx+1)
    sns.scatterplot(x=f, y='SalePrice', data=df)
    plt.grid(1)
```

Now we can infer that, 
* The price decreased as the age of the property increased. Still, a few propeties were sold at high price maybe because they we antique or something (most prob)
* Same happened with Remodification Year and Garage Built Year

**PREPROCESSING** Thus, we'll replace the year number with the year count.

```python
for f in temporal_num_features:
    data[f] = data['YrSold'] - data[f]

# Now that we've merged the YrSold information, we don't need this feature anymore.
data.drop(columns=['YrSold'], inplace=True)
data.head()
```

#### Discrete and Cont Numerical Features

```python
num_cont_features     = []
num_discrete_features = []

for f in num_features:
    if f not in temporal_num_features:
        if len(data[f].unique()) > 25:
            num_cont_features.append(f)
        else:
            num_discrete_features.append(f)

print('Total number of Continuous Numerical Features :', len(num_cont_features))
print('Total number of Discrete Numerical Features   :', len(num_discrete_features))
```

```python
plt.figure(figsize=(30, 15))
for idx, f in enumerate(num_cont_features):
    
    plt.subplot(4, 5, idx+1)
    data[f].plot(kind='hist', bins=40)
    plt.grid(1)
    plt.title(f)
```

As we can see, the data is highly skewed, so, we'll use log normalization to fix this towards Normalization.

```python
plt.figure(figsize=(15, 7))

i = 0
for f in num_cont_features:
    
    if 0 not in data[f].unique(): # Cannot apply log to 0 vals.
        
        plt.subplot(2, 4, i+1)
        np.log(data[f]).plot(kind='hist', bins=40)
        plt.grid(1)
        plt.title(f)
        i += 1
```

```python
df = data.copy()

for f in num_cont_features:
    df[f] = np.where(df[f] == 0, df[f].mean(), df[f])
```

```python
plt.figure(figsize=(15, 7))

i = 0
for f in num_cont_features:

    plt.subplot(5, 4, i+1)
    np.log(df[f]).plot(kind='hist', bins=40)
    plt.grid(1)
    plt.title(f)
    i += 1
```

```python
# def predict():
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


features = data.drop(columns=['SalePrice'])
target   = data['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=10)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)


print('RMSE :', np.sqrt(mean_squared_error(np.log(y_test), np.log(pred))))
print('R2   :', r2_score(y_test, pred))
```

```python
model = GradientBoostingRegressor(learning_rate=0.2, n_estimators=170)
model.fit(X_train, y_train)
pred = model.predict(X_test)


print('RMSE :', np.sqrt(mean_squared_error(np.log(y_test), np.log(pred))))
print('R2   :', r2_score(y_test, pred))
```

```python
def feature_eng(data):
    
    from sklearn.preprocessing import LabelEncoder
    
    THRESHOLD = 80
    nan_features = data.columns[data.isnull().sum() > 1]
    nan_percentage_of_features = data[nan_features].isnull().sum()/len(data)*100
    features_higher_than_threshold_nan = nan_percentage_of_features[nan_percentage_of_features>THRESHOLD]
    data = data.drop(columns=features_higher_than_threshold_nan.index)


    data['FireplaceQu'].fillna('missing', inplace=True)
    data['LotFrontage'].fillna(data['LotFrontage'].median(), inplace=True)

    cat_features = [f for f in data.columns if data[f].dtype=='O']
    num_features = [f for f in data.columns if data[f].dtype!='O' and f not in ['Id', 'SalePrice']]
    
    THRESHOLD = 5
    df = data.copy()
    for f in cat_features:
        df[f].fillna('other', inplace=True)
        frequency_percent = df.groupby(f)[f].count()/len(data) * 100
    #     print(frequency_percent, '\n')
        for val in frequency_percent[frequency_percent<THRESHOLD].index:
            df[f] = np.where(df[f] == val, 'other', df[f])
    data = df.copy()
    
    
    data['MasVnrArea'].fillna(data['MasVnrArea'].median(), inplace=True)
    data['GarageYrBlt'].fillna(data['GarageYrBlt'].mean(), inplace=True)

    biased_features = ['Street', 'Utilities', 'LandSlope', 'Condition2', 'RoofMatl', 'CentralAir', 'Heating', 'Electrical', 'Functional', 'GarageCond', 'GarageQual', 'PavedDrive', 'SaleType', 'SaleCondition']

    biased_features_with_no_relation = [f for f in biased_features if f not in ['CentralAir', 'GarageCond', 'GarageQual', 'SaleType']]
    data.drop(columns=biased_features_with_no_relation, inplace=True)

    cat_features = [f for f in data.columns if data[f].dtype=='O']
    num_features = [f for f in data.columns if data[f].dtype!='O' and f not in ['Id', 'SalePrice']]
    
#     return data
    encoder = LabelEncoder()
    for f in cat_features:
        data[f] = encoder.fit_transform(data[f])


    temporal_num_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']

    for f in temporal_num_features:
        data[f] = data['YrSold'] - data[f]

    # Now that we've merged the YrSold information, we don't need this feature anymore.
    data.drop(columns=['YrSold'], inplace=True)
    
    num_cont_features     = []
    num_discrete_features = []
    for f in num_features:
        if f not in temporal_num_features:
            if len(data[f].unique()) > 25:
                num_cont_features.append(f)
            else:
                num_discrete_features.append(f)
          
        
    num_features = [f for f in data.columns if data[f].dtype!='O' and f not in ['Id', 'SalePrice']]
    for f in num_features:
        data[f].fillna(data[f].median(), inplace=True)
        
    return data
```
